```python
import numpy as np
import sklearn as sk
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')
np.random.seed(2)
```

```python
import matplotlib.pyplot as plt
import time
from sklearn import model_selection as ml
```

```python
split_data = mnist.data[0:4900]
split_target = mnist.target[0:4900]

xtrain, xtest, ytrain, ytest = ml.train_test_split(split_data, split_target, train_size=0.7)
```

```python
from sklearn.svm import SVC

print(xtrain.shape)
print(ytrain.shape)

L = ['linear', 'poly', 'rbf', 'sigmoid']#, 'precomputed']
res = []
temps = []

for k in L:
    print("test with kernel: ", k)
    clf = SVC(kernel = k, gamma = 'auto')
    start = time.time()
    clf.fit(xtrain, ytrain)
    stop = time.time()
    
    res.append(clf.score(xtest, ytest))
    temps.append(stop-start)
```

```python
plt.plot(L, temps)
plt.show()

plt.plot(L, res)
plt.show()
```

```python
from sklearn.svm import SVC

print(xtrain.shape)
print(ytrain.shape)

L = np.linspace(0.1, 1, 5)
res = []
res_train = []
temps = []

for k in L:
    print("test with c: ", k)
    clf = SVC(kernel = 'poly', gamma = 'auto', C = k)
    start = time.time()
    clf.fit(xtrain, ytrain)
    stop = time.time()
    
    res.append(clf.score(xtest, ytest))
    res_train.append(clf.score(xtrain, ytrain))
    temps.append(stop-start)
```

```python
plt.plot(L, temps)
plt.show()

plt.plot(L, res, label="test data score")
plt.plot(L, res_train, label="train data score")
plt.legend()
plt.show()
```

## Comparaison des 3 methodes avec les meilleurs hyperparametres pour chacun d'elle

```python
from sklearn import model_selection as ml
from sklearn import neighbors
from sklearn.neural_network import MLPClassifier as mlp
```

```python
def k_fold(X, Y, n, clf):
    res = []
    kf = ml.KFold(n_splits=n, shuffle=True)
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        clf.fit(X_train, Y_train)
        
        res.append(clf.score(X_test, Y_test))
        
        return np.mean(res)
```

```python
#test des neighbors avec knn sur ces 4900 data pour selctionner le plus opti
n_folds = 10
L = []
G = []
H = []
for k in range(2,16):
    clf = neighbors.KNeighborsClassifier(k)
    start = time.time()
    res = k_fold(split_data, split_target, n_folds, clf)
    stop = time.time()
    L.append(k)
    G.append(res)
    H.append(stop-start)
    print("k: ", k, "  , score: ", res)
    
plt.plot(L,G)
plt.show()

plt.plot(L,H)
plt.show()
```

```python
meth = ['knn', 'mlp', 'svc']
temps_fit = []
temps_train = []
temps_score = []
preci_test = []
preci_train = []
cm = []
```

```python
#KNN
clf = neighbors.KNeighborsClassifier(n_neighbors = 3)

start = time.time()
clf.fit(xtrain, ytrain)
stop = time.time()
temps_fit.append(stop-start)

start = time.time()
preci_test.append(clf.score(xtest, ytest))
stop = time.time()
temps_score.append(stop-start)

start = time.time()
preci_train.append(clf.score(xtrain, ytrain))
stop = time.time()
temps_train.append(stop-start)

cm.append(confusion_matrix(ytest, clf.predict(xtest)))
```

```python
#MLP
couches = np.linspace(11, 60, 10, dtype=int)
couches = couches.tolist()
couches.reverse()
couches = tuple(couches)

clf = mlp(hidden_layer_sizes = couches)

start = time.time()
clf.fit(xtrain, ytrain)
stop = time.time()
temps_fit.append(stop-start)

start = time.time()
preci_test.append(clf.score(xtest, ytest))
stop = time.time()
temps_score.append(stop-start)

start = time.time()
preci_train.append(clf.score(xtrain, ytrain))
stop = time.time()
temps_train.append(stop-start)

cm.append(confusion_matrix(ytest, clf.predict(xtest)))
```
