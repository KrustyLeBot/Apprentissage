```python
import numpy as np
import sklearn as sk
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')
np.random.seed(2)
```

```python
import matplotlib.pyplot as plt
import time
from sklearn import model_selection as ml
```

```python
split_data = mnist.data[0:4900]
split_target = mnist.target[0:4900]

xtrain, xtest, ytrain, ytest = ml.train_test_split(split_data, split_target, train_size=0.7)
```

```python
from sklearn.neural_network import MLPClassifier as mlp

clf = mlp(hidden_layer_sizes = (50))
clf.fit(xtrain, ytrain)
```

```python
print("Précision du classifieur: ", clf.score(xtest, ytest))
```

```python
x = 4
images = xtest[x].reshape((28,28))
plt.imshow(images, cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
print("Target associé: ", ytest[x])
print("Prédiction de l'algo: ", clf.predict(xtest[x].reshape(1, -1)))
```

```python
from sklearn.metrics import precision_score
```

```python
xtest_predict = clf.predict(xtest)

res = precision_score(ytest, xtest_predict, average = 'micro')
print(res)
```

# Reseau de neurones sans degression entre les couches - defaut

```python
preci = []
liste = [2,10,20,50,100]
temps = []


for n in liste:
    couches = 50*np.ones((n,), dtype=int)
    couches = tuple(couches.tolist())
    clf = mlp(hidden_layer_sizes = couches)
    print("fit nb couches: ", n)
    start = time.time()
    clf.fit(xtrain, ytrain)
    stop = time.time()
    preci.append(clf.score(xtest, ytest))
    temps.append(stop-start)
```

```python
plt.plot(liste, temps)
plt.show()

plt.plot(liste, preci)
plt.show()
```

# Reseau de neurones avec degression entre les couches - defaut

```python
preci = []
debut = 60
fin = 11

liste = [2,10,20,50]
temps = []


for n in liste:
    print("fit nb couches: ", n)
    couches = np.linspace(fin, debut, n, dtype=int)
    couches = couches.tolist()
    couches.reverse()
    couches = tuple(couches)
    print("Couches: ", couches)
    clf = mlp(hidden_layer_sizes = couches)
    start = time.time()
    clf.fit(xtrain, ytrain)
    stop = time.time()
    preci.append(clf.score(xtest, ytest))
    temps.append(stop-start)
```

```python
plt.plot(liste, temps)
plt.show()

plt.plot(liste, preci)
plt.show()
```

# Reseau de neurones sans degression entre les couches

```python
preci = [[],[],[]]
liste = [2,10,20,50,100]
opti = ['adam', 'lbfgs', 'sgd']
temps = [[],[],[]]

for sol in range(len(opti)):
    solv = opti[sol]
    for n in liste:
        couches = 50*np.ones((n,), dtype=int)
        couches = tuple(couches.tolist())
        clf = mlp(hidden_layer_sizes = couches, solver = solv)
        print("fit nb couches: ", n)
        start = time.time()
        clf.fit(xtrain, ytrain)
        stop = time.time()
        preci[sol].append(clf.score(xtest, ytest))
        temps[sol].append(stop-start)
```

```python
for z in range(len(opti)):
    plt.plot(liste, temps[z], label=opti[z])
    plt.legend()
plt.show()

for z in range(len(opti)):
    plt.plot(liste, preci[z], label=opti[z])
    plt.legend()
plt.show()
```

# Model le plus opti: Adam avec 10 couches degressif de 60 a 11 neurone

```python
from sklearn.metrics import zero_one_loss

couches = np.linspace(11, 60, 10, dtype=int)
couches = couches.tolist()
couches.reverse()
couches = tuple(couches)

L = ['relu', 'identity', 'logistic', 'tanh']
res = []
temps = []
zero_one = []

for k in L:
    print("fit activation: ", k)
    
    clf = mlp(hidden_layer_sizes = couches, activation = k)
    start = time.time()
    clf.fit(xtrain, ytrain)
    stop = time.time()
    x = clf.score(xtest, ytest)
    print(x)
    res.append(clf.score(xtest, ytest))
    zero_one.append(zero_one_loss(ytest, clf.predict(xtest), normalize=True))
    temps.append(stop-start)
```

```python
plt.plot(L, temps)
plt.show()

plt.plot(L, res)
plt.show()

plt.plot(L, zero_one)
plt.show()
```

```python
couches = np.linspace(11, 60, 10, dtype=int)
couches = couches.tolist()
couches.reverse()
couches = tuple(couches)

L = [10**(-3), 10**(-4), 10**(-5), 10**(-6)]
L_str = ["10**(-3)", "10**(-4)", "10**(-5)", "10**(-6)"]
res = []
temps = []
zero_one = []

for k in L:
    print("alpha: ", k)
    
    clf = mlp(hidden_layer_sizes = couches, alpha = k)
    start = time.time()
    clf.fit(xtrain, ytrain)
    stop = time.time()
    x = clf.score(xtest, ytest)
    print(x)
    res.append(clf.score(xtest, ytest))
    zero_one.append(zero_one_loss(ytest, clf.predict(xtest), normalize=True))
    temps.append(stop-start)
```

```python
plt.plot(L_str, temps)
plt.show()

plt.plot(L_str, res)
plt.show()

plt.plot(L_str, zero_one)
plt.show()
```
