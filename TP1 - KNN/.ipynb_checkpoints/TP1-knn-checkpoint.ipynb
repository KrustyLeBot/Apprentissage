```python
import numpy as np
import sklearn as sk
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')
np.random.seed(2)
print("done")
```

```python
import matplotlib.pyplot as plt
import time
```

```python
images = mnist.data.reshape((-1,28,28))
x = 2000
plt.imshow(images[x], cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
print("Target associé: ", mnist.target[x])
```

```python
from sklearn import model_selection as ml
from sklearn import neighbors
```

```python
data = np.random.randint(70000, size=5000)
split_data = mnist.data[data]
split_target = mnist.target[data]
xtrain, xtest, ytrain, ytest = ml.train_test_split(split_data, split_target, train_size=0.8)
```

```python
n_neighbors = 10
clf = neighbors.KNeighborsClassifier(n_neighbors)
```

```python
clf.fit(xtrain, ytrain)
```

```python
y = 6
image = split_data[y].reshape(28,28)
test = split_data[y].reshape(1, -1)
print(clf.predict(test))
plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

```python
print(clf.score(xtest, ytest))
```

Il est normal de ne pas avoir un taux de reussite de 100%.

```python
def k_fold(X, Y, n, clf):
    res = []
    kf = ml.KFold(n_splits=n, shuffle=True)
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        clf.fit(X_train, Y_train)
        
        res.append(clf.score(X_test, Y_test))
        
        return np.mean(res)
```

```python
n_folds = 10
L = []
G = []
H = []
for k in range(2,16):
    clf = neighbors.KNeighborsClassifier(k)
    start = time.time()
    res = k_fold(split_data, split_target, n_folds, clf)
    stop = time.time()
    L.append(k)
    G.append(res)
    H.append(stop-start)
    print("k: ", k, "  , score: ", res)
    
plt.plot(L,G)
plt.show()

plt.plot(L,H)
plt.show()
```

Le meilleur est k = 3

```python
LL = []
GG = []
HH = []
for i in range(1,10):
    xtrain, xtest, ytrain, ytest = ml.train_test_split(split_data, split_target, train_size=(i/10.0))
    clf = neighbors.KNeighborsClassifier(n_neighbors = 3)
    start = time.time()
    clf.fit(xtrain, ytrain)
    stop = time.time()
    z = clf.score(xtest, ytest)
    LL.append(i)
    GG.append(z)
    HH.append(stop-start)
    print("train size: ", (i/10.0), "%  , score: ", z)
    
plt.plot(LL, HH)
plt.show()

plt.plot(LL, GG)
plt.show()
```

On voit clairement que le pourcentage permettant d'avoir les meilleur score(sans kfold) est: 0.9%
Plus on a d'image plus on a de chance de faire un bon score car l'algo s'entraine mieux. Mais, il est a noter, que plus l'echantillon de test est petit, plus il est difficile de faire confiance au score donne

```python
L = [100, 1000, 10000, 30000]
for j in L:
    data = np.random.randint(70000, size=j)
    split_data = mnist.data[data]
    split_target = mnist.target[data]
    
    xtrain, xtest, ytrain, ytest = ml.train_test_split(split_data, split_target, train_size=0.8)
    clf = neighbors.KNeighborsClassifier(n_neighbors = 3)
    clf.fit(xtrain, ytrain)
    print("Dataset size: ", j, "  , score: ", clf.score(xtest, ytest))
```

On se base sur un pourcentage de 0.8 pour train et 0.2 pour test et un k de 3 car c'etait le meilleur dans l'evaluation precedente. On voit clairement que:
- Plus le dataset est grand plus l'algo met du temps a s'entrainer
- Plus le dataset est grand, plus l'algo est precis, mais l'amelioration n'est clairement pas lineaire

```python
K = [1, 2]
data = np.random.randint(70000, size=10000)
split_data = mnist.data[data]
split_target = mnist.target[data]
xtrain, xtest, ytrain, ytest = ml.train_test_split(split_data, split_target, train_size=0.8)
for j in K:
    clf = neighbors.KNeighborsClassifier(n_neighbors = 3, p = j)
    clf.fit(xtrain, ytrain)
    print("p: ", j, "  , score: ", clf.score(xtest, ytest))
```

p = 1 : Minkowski distance
p = 2 : Euclidian distance / default distance used

On se base sur un echantillon de 10000, pour avoir a la fois un resultat assez precis, mais egallement un temps de calculs pas trop long; k=3 et train_size=0.8%. On voit la distance euclidienne est un peu plus performante, mais de peu.

```python
K = [1, -1]
data = np.random.randint(70000, size=5000)
split_data = mnist.data[data]
split_target = mnist.target[data]
xtrain, xtest, ytrain, ytest = ml.train_test_split(split_data, split_target, train_size=0.8)
for j in K:
    
    clf = neighbors.KNeighborsClassifier(n_neighbors = 3, n_jobs = j)
    clf.fit(xtrain, ytrain)
    
    start = time.time()
    for ww in range(10):
        clf.score(xtest, ytest)
        print("done:", ww)
    end = time.time()
    
    print("n_jobs: ", j, "  , score: ", clf.score(xtest, ytest), " , time used: ", (end - start))
```

n_jobs = 1 : 1 parallel job
n_jobs = -1 : all processor is used

On realise une boucle de 100 fit, puis on fait la moyenne des temps
Tres peu de difference entre les deux methodes. Sur un PC tres performant, la difference sera peut etre plus marquee.

Avantage de K-nn:
- training phase tres rapide
- il est innutile d'entrainer un model pour faire de la generalisation
- utilisable pour des donnees non lineaires

Inconvenients de K-nn:
- testing phase lente et requiert beaucoup de memoir pour mettre le dataset en entier dedans
- pas fait pour des datasets tres larges
- tres sensible au grandeur des distances
